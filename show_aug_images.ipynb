```python
import os
import argparse
import statistics

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from PIL import Image
from collections import defaultdict
```

```python
def show_all(src, image_files: list[str], rows, cols):
    '''
    shows a subplot for each image file    
    '''
    
    fig, axs = plt.subplots(rows, cols)

    idx = 0
    for i in range(rows):
        for j in range(cols):
            filename = image_files[idx]
            img = mpimg.imread(f'{src}/{filename}')
            axs[i][j].imshow(img)
            axs[i][j].tick_params(left = False, right = False , labelleft = False , 
                        labelbottom = False, bottom = False)
            
            idx += 1
            
    plt.show()
```

```python
src = 'training/oraisisaac'
image_files = [file for file in os.listdir(src) if file.split('_')[1] == '0111']
len(image_files[:20])
```

```python
show_all(src, image_files[:18], 3, 6)
show_all(src, image_files[18:], 3, 6)
```
